# step5: Formulate Hypotheses

# Do males and females spend differently?

```python
# H0 = Males and females spending are same. 
# H1 = Males and females spending are not same.
```

```python
male = df[df['Gender'] == 'Male']['MonthlySpend']
female = df[df['Gender'] == 'Female']['MonthlySpend']
```

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(male, female)
print(t_stat, p_value)
```

```python
''' Here p_value >0.05 so we will take H0.(Males and females spending are same.).'''
```

# Does education level impact average monthly spend?

```python
#H0: Education level has no effect on spending.
#H1: Education level affects average monthly spending.
```

```python
from scipy.stats import f_oneway

groups = [group['MonthlySpend'].values for name, group in df.groupby('Education')]
f_stat, p_value = f_oneway(*groups)
print(f_stat)
print(p_value)
```

```python
'''Since p-value > 0.05, we fail to reject H₀. Education level has no significant effect on spending.'''
```

# Does state-wise spend vary significantly?

```python
df["State"].unique()
```

```python
#H0:Average spending is same across all states.
#H1:At least one state has different average spending.
```

```python
from scipy.stats import f_oneway
groups = [group['MonthlySpend'].values for name, group in df.groupby('State')]
f_stat, p_value = f_oneway(*groups)

print("F-stat:", f_stat)
print("p-value:", p_value)
```

```python
if p_value >0.05:
    print("Average spending is same across all states.")
elif p_value<0.05:
    print("At least one state has different average spending.")
```

# step6

```python
1. Education vs Monthly Spend:

H₀ (Null Hypothesis): Education level has no significant effect on Monthly Spend.
H₁ (Alternative Hypothesis): Education level has a significant effect on Monthly Spend.
```

```python
2. Gender vs Monthly Spend:

H₀ (Null Hypothesis): There is no significant difference in average Monthly Spend between males and females.
H₁ (Alternative Hypothesis): There is a significant difference in average Monthly Spend between males and females.
```

```python
3. State vs Monthly Spend:

H₀ (Null Hypothesis): The average Monthly Spend is the same across all states.
H₁ (Alternative Hypothesis): At least one state has a different average Monthly Spend.
```

```python
- Gender vs Monthly Spend → Independent t-test
- Education vs Monthly Spend → One-way ANOVA
- State vs Monthly Spend → One-way ANOVA
```

```python
Check Assumptions

Normality was checked using the Shapiro-Wilk test. 
Independence was assumed since each observation represents a different customer.
Homogeneity of variance was tested using Levene’s test. 
If p-value > 0.05, the respective assumption (normality or equal variance) is satisfied.
```

```python
Interpretation of p-values and Confidence Intervals:

If p-value < 0.05, reject the null hypothesis (significant result). 
If p-value > 0.05, do not reject the null hypothesis (no significant result). 
Confidence intervals show the range of the true value; if it does not include the null value, the result is significant.
```
